```python
#read the file 
train = pd.read_csv("train.csv")
```

```python
#pare tis suntetagmenes
game=list(zip(train.latitudeF,train.longitudeF,train.room_typeF,train.transitF))

#xrisimopoiise mono 100 sintetagmenes
game1=game[0:100]
```

```python
#map for 100 apartment 
import folium
#syntetagmenes athens
m = folium.Map(
    location=[37.983810,  23.727539],
    zoom_start=12,
    tiles='Stamen Terrain'
)

tooltip = 'Click me!'

for coord in game1:
    folium.Marker([ coord[0], coord[1] ], popup=folium.Popup(coord[2]),tooltip=tooltip).add_to(m)

m
```
