```python
#read the file 
train = pd.read_csv("train.csv")
```

```python
#create list and arraywith id and description
df=list(zip(train.idF,train.descriptionF))
df1=pd.DataFrame(data = df, columns=['id','descriptionF'])
df=list(zip(train.idM,train.descriptionM))
df2=pd.DataFrame(data = df, columns=['id','descriptionM'])
df=list(zip(train.idA,train.descriptionA))
df3=pd.DataFrame(data = df, columns=['id','descriptionA'])

#merge base in id
df4=pd.merge(df1, df2,how='outer')
df4=pd.merge(df4, df3,how='outer')

#create array withh description from  merge
df=list(zip(df4.descriptionF,df4.descriptionM,df4.descriptionA))
dfall=pd.DataFrame(data=df,columns=['descriptionF','descriptionM','descriptionA'])


#new list and array with description and id
df=list(zip(df4.id,df4.descriptionF))
df1=pd.DataFrame(data = df, columns=['id','description'])


df=list(zip(df4.id,df4.descriptionM))
df2=pd.DataFrame(data = df, columns=['id','description'])
#df2

df=list(zip(df4.id,df4.descriptionA))
df3=pd.DataFrame(data = df, columns=['id','description'])
#df3

#sinxwneusi gia na fugoun ta nan kai na exw ena sunoliko xwris dipla description kai id 
df4 = df1.combine_first(df2)
dall=df3.combine_first(df4)

dall = dall[dall['description'].notna()]

dall=dall.head(200)
```

```python
#ERWTIMA 3.1
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(dall.description )
#print(vectorizer.get_feature_names())
uni_dig=tfidf.toarray()
uni_dig
```

```python
#ERWTIMA 3.2
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(uni_dig)
df3=pd.DataFrame(data = cosine_sim)
```
