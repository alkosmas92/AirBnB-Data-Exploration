```python
#read the file
train = pd.read_csv("train.csv")
```

```python
#create list and array with neighbourhood and one column One with every line have 1,for every month
most_freq=list(zip(train.neighbourhoodF))
df1 = pd.DataFrame(data = most_freq, columns=['neighbourhoodF'])
df1['One']=1
               
most_freq=list(zip(train.neighbourhoodM))
df2 = pd.DataFrame(data = most_freq, columns=['neighbourhoodF'])
df2['One']=1

               
most_freq=list(zip(train.neighbourhoodF))
df3 = pd.DataFrame(data = most_freq, columns=['neighbourhoodF'])
df3['One']=1

#create list and array with neighbourhood and room_type and One
most_freq=list(zip(train.neighbourhoodF,train.room_typeF,df1.One))
df1 = pd.DataFrame(data = most_freq, columns=['neighbourhood','room_type','One'])

most_freq=list(zip(train.neighbourhoodM,train.room_typeM,df2.One))
df2 = pd.DataFrame(data = most_freq, columns=['neighbourhood','room_type','One'])

most_freq=list(zip(train.neighbourhoodA,train.room_typeA,df3.One))
df3 = pd.DataFrame(data = most_freq, columns=['neighbourhood','room_type','One'])
```

```python
#find the count for every room_type in every neighbourhood per month
df1 = pd.DataFrame(data = df1.groupby(['neighbourhood','room_type']).sum())
df2 = pd.DataFrame(data = df2.groupby(['neighbourhood','room_type']).sum())
df3 = pd.DataFrame(data = df3.groupby(['neighbourhood','room_type']).sum())
```

```python
#do array again
ddf_1=df1.reset_index()
ddf_2=df2.reset_index()
ddf_3=df3.reset_index()
```

```python
#kartesiano
df_row = pd.concat([ddf_1,ddf_2,ddf_3])
```

```python
#find the count for every room_type in every neighbourhood all month

dff2 = pd.DataFrame(data = df_row.groupby(['neighbourhood','room_type']).sum())
```

```python
#diagram for the count every room_type in every neighbourhood all month
dff2['One'].plot.bar(figsize=(20, 4))

print("The most popular name")
dff2.sort_values(by='One', ascending=False)
```
