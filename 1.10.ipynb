```python
train = pd.read_csv("train.csv")
#train
```

```python
df=list(zip(train.idF,train.transitF))
df1=pd.DataFrame(data = df, columns=['id','transitF'])
df=list(zip(train.idM,train.transitM))
df2=pd.DataFrame(data = df, columns=['id','transitM'])
df=list(zip(train.idA,train.transitA))
df3=pd.DataFrame(data = df, columns=['id','transitA'])            #Dimiourgia pinaka me id kai kathe transit 
                                                                  #gia kathe id ta 3 transit tou
df4=pd.merge(df1, df2,how='outer')
df4=pd.merge(df4, df3,how='outer')
df4
                                                                            
df=list(zip(df4.transitF,df4.transitM,df4.transitA))
dfall=pd.DataFrame(data=df,columns=['transitF','transitM','transitA'])
dfall


df=list(zip(df4.id,df4.transitF))
df1=pd.DataFrame(data = df, columns=['id','transit'])             #pianakas id-transit me feb
#df1

df=list(zip(df4.id,df4.transitM))
df2=pd.DataFrame(data = df, columns=['id','transit'])            #pinakas id-transit me march
#df2

df=list(zip(df4.id,df4.transitA))
df3=pd.DataFrame(data = df, columns=['id','transit'])          #pinakas id-transit me april
#df3

df4 = df1.combine_first(df2)
dall=df3.combine_first(df4)
dall                                                        #combine pinakas

dall = dall[dall['transit'].notna()]                         #aferesi NaN
#dall
```

```python
import pandas as pd
dall=list(zip(dall.transit))
dall=pd.DataFrame(dall,columns=['text'])
dall=dall.reset_index()                                             #To results meta to join tha xrisimopoihthei 
                                                                    #gia to wordcould
results = set()
dall['text'].str.lower().str.split().apply(results.update)
results=list(results)
results=' '.join(results)
#results
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create the wordcloud object
#wordcloud = WordCloud(width=480, height=480, margin=0).generate(results)
 
wordcloud = WordCloud(max_words=100,
                          background_color="white",
                          width=1024,
                          height=768).generate(results)
  
    
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
# Create the wordcloud object
#wordcloud = WordCloud(width=480, height=480, margin=0).generate(results)
counts = dall['text'].value_counts()

counts.index = counts.index.map(str)

from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud().generate_from_frequencies(counts)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
